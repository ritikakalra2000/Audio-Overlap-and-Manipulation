### Task 1

For task 1 we are using library pydub and Ipython.display for listening to the audio.<br>
We combine 2 sounds - audio1 and audio2 and put the resultant audio in result.

```python
from pydub import AudioSegment
import parselmouth
from IPython.display import Audio
```

```python
sound1 = AudioSegment.from_file("audio_1.wav")
sound2 = AudioSegment.from_file("audio_2.wav")
combined = sound1.overlay(sound2)
combined.export("res.wav", format='wav')
```

```python
sound1 = parselmouth.Sound("audio_1.wav")
Audio(sound1,rate=sound1.sampling_frequency)
```

```python
sound2 = parselmouth.Sound("audio_2.wav")
Audio(data=sound2.values, rate=sound2.sampling_frequency)
```

```python
result = parselmouth.Sound("res.wav")
Audio(data=result.values, rate=result.sampling_frequency)
```

### Task 2

For Task 2, we are using library parselmouth.praat to manipulate the male voice to female voice

```python
sound = parselmouth.Sound("nice-work.wav")
Audio(data=sound.values, rate=sound.sampling_frequency)
```

```python
from parselmouth.praat import call
manipulation = call(sound, "To Manipulation", 0.05, 100, 1100)
pitch_tier = call(manipulation, "Extract pitch tier")

call(pitch_tier, "Multiply frequencies", sound.xmin, sound.xmax, 3)

call([pitch_tier, manipulation], "Replace pitch tier")
sound_octave_up = call(manipulation, "Get resynthesis (overlap-add)")
Audio(data=sound_octave_up.values, rate=sound_octave_up.sampling_frequency)
```
